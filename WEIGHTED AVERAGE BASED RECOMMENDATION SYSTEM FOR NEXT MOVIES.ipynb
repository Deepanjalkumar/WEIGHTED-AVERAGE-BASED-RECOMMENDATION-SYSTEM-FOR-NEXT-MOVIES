```python
movie=pd.read_csv(r"C:\Users\Deepanjal\Downloads\tmdb_5000_movies.csv.zip")
movie
```

```python
credit=credit.rename(columns={'movie_id':'id'})
```

```python
credit.head(1)
```

```python
#combine
data=movie.merge(credit,on='id',how='inner')
```

```python
data
```

```python
data=data.drop(columns={'homepage','title_x','title_y','status','production_countries'})
```

```python
data.info(null_counts=True)
```

```python
v=data['vote_count']
R=data['vote_average']
C=data['vote_average'].mean()
m=data['vote_count'].quantile(0.70)
```

```python
data['weighted_average']=((R*v)+(C*m))/(v+m)
```

```python
data["weighted_average"]
```

```python
movie_sorted_ranking=data.sort_values('weighted_average',ascending=False)
movie_sorted_ranking[['original_title','vote_count','vote_average','weighted_average','popularity']].head(20)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
movie_sorted_ranking=data.sort_values('popularity',ascending=False)
```

```python
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=movie_sorted_ranking['weighted_average'].head(10),y=movie_sorted_ranking['original_title'].head(10))
plt.xlim(4,10)
plt.title('Best Movies By Average Votes',weight='bold')
plt.xlabel('weighted average score', weight='bold')
plt.ylabel('movie title',weight='bold')
plt.savefig('best_mmovies.png')
```

```python
#based upon popularity  and weighted average in the ratio of 0.5 and 0.5
from sklearn.preprocessing import MinMaxScaler
scaling=MinMaxScaler()
movie_scaled_df=scaling.fit_transform(data[['weighted_average','popularity']])
movie_normalized_df=pd.DataFrame(movie_scaled_df,columns=['weighted_average','popularity'])
movie_normalized_df.head()
```

```python
data[['normalized_weight_average','normalized_popularity']]= movie_normalized_df
```

```python
data['score'] = data['normalized_weight_average'] * 0.5 + data['normalized_popularity'] * 0.5
movies_scored_df = data.sort_values(['score'], ascending=False)
movies_scored_df[['original_title', 'normalized_weight_average', 'normalized_popularity', 'score']].head(20)
```

```python
scored_df = data.sort_values('score', ascending=False)
plt.figure(figsize=(16,6))
ax = sns.barplot(x=scored_df['score'].head(10), y=scored_df['original_title'].head(10), data=scored_df, palette='deep')
#plt.xlim(3.55, 5.25)
plt.title('Best Rated & Most Popular Blend', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')

plt.savefig('scored_movies.png')
```
